```python
# Redesign plotting
import matplotlib.pyplot as plt

model_type = ['Baseline', 'VICGAE', 'Chemprop', 'Chemeleon', 'SMI-TED']
label_type = ['Fingerprint', 'VICGAE', 'Chemprop', 'Chemeleon', 'SMI-TED']

# Extract the RMSE lists from results dictionary
auc_lists = [results_dict1[model]['RMSE'].values for model in model_type]

plt.figure(figsize=(8, 5))
box = plt.boxplot(auc_lists, patch_artist=True, labels=model_type)

for patch in box['boxes']:
    patch.set_facecolor('#1f77b4')
    patch.set_alpha(0.6)

plt.ylabel('RMSE', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(fontsize=16, rotation=15)
plt.yticks(fontsize=16)
plt.xticks(ticks=range(1, len(label_type) + 1), labels=label_type, fontsize=16, rotation=15)

# Refinements
plt.ylim(bottom=0.3)                  
plt.tick_params(axis='x', top=True, labeltop=True, bottom=False, labelbottom=False)  # move xticks to top
plt.xlabel('')                       

plt.tight_layout()

# plt.savefig("../../results/surfactants_model_performance.png", dpi=600, bbox_inches="tight")
plt.show()
```
