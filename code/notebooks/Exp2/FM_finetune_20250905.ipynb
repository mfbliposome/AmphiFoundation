```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Example: x_latent.shape = (50, 768)
n_subsets = 5
subset_size = 10
labels = [
    "high_replace3_dissimilar_finetune",
    "high_replace3_dissimilar",
    "low_replace5_smi_finetune",
    "low_replace5_smi",
    "high_replace5_smi_finetune"
]

# Assign group labels
group_labels = []
for i in range(n_subsets):
    group_labels.extend([labels[i]] * subset_size)
group_labels = np.array(group_labels)

# Perform hierarchical clustering
Z = linkage(x_latent, method="ward")

# Assign subset colors
subset_colors = {
    "high_replace3_dissimilar_finetune": "tab:blue",
    "high_replace3_dissimilar": "tab:orange",
    "low_replace5_smi_finetune": "tab:green",
    "low_replace5_smi": "tab:red",
    "high_replace5_smi_finetune": "tab:purple"
}

# Map sample index -> color
sample_colors = {i: subset_colors[group_labels[i]] for i in range(len(group_labels))}

# --- Build cluster membership mapping ---
from collections import defaultdict

n_samples = len(group_labels)
cluster_to_members = defaultdict(list)

# Initialize leaves
for i in range(n_samples):
    cluster_to_members[i] = [i]

# Fill in internal nodes from linkage
for cluster_id, (c1, c2, _, _) in enumerate(Z, start=n_samples):
    cluster_to_members[cluster_id] = cluster_to_members[int(c1)] + cluster_to_members[int(c2)]

# Function to decide branch color
def link_color_func(node_id):
    members = cluster_to_members[node_id]
    member_labels = group_labels[members]
    unique_labels = np.unique(member_labels)
    if len(unique_labels) == 1:
        return subset_colors[unique_labels[0]]  # homogeneous branch
    else:
        return "black"  # mixed branch

# --- Plot dendrogram ---
plt.figure(figsize=(12, 6))
dendrogram(
    Z,
    labels=group_labels,
    leaf_rotation=90,
    leaf_font_size=10,
    link_color_func=link_color_func
)

# Color the tick labels
ax = plt.gca()
xlbls = ax.get_xmajorticklabels()
for lbl in xlbls:
    lbl.set_color(subset_colors[lbl.get_text()])

plt.ylabel("Distance (Ward linkage)")
plt.title("Hierarchical Clustering of Latent Space Samples")
plt.tight_layout()
plt.show()
```

```python
from tqdm import tqdm
import numpy as np

def evaluate_model(model, replacement_sets, target=100):
    info_path = '/Users/haoliu/Documents/Hao/Fordham/Cheminformatics/MD/Amphiphiles/info_new_molecules.csv'
    df_info = pd.read_csv(info_path)
    new_smiles = df_info['SMILES'].iloc[:5].tolist()
    
    pbar = tqdm(total=target, desc="High-prob samples found", unit="sample")
    high_prob_dfs = []
    total_found = 0
    samples_screened = 0  # Track how many samples were screened

    for replacement_set in replacement_sets:
        df_new = df_large.copy()
        for col, new_smi in zip(replacement_set, new_smiles):
            df_new[col] = new_smi

        x_new = get_latent_space(df_new)
        y_pred = model.predict_proba(x_new)[:, 1]
        df_new['vesicles_formation'] = y_pred

        high_prob_indices = np.where(y_pred > 0.8)[0]
        samples_screened += len(df_new)  # We "looked at" all rows

        if len(high_prob_indices) > 0:
            high_samples = df_new.iloc[high_prob_indices].copy()

            remaining = target - total_found
            if len(high_samples) > remaining:
                high_samples = high_samples.iloc[:remaining]

            high_prob_dfs.append(high_samples)
            pbar.update(len(high_samples))
            total_found += len(high_samples)

            if total_found >= target:
                break

    pbar.close()
    return samples_screened, high_prob_dfs
```
