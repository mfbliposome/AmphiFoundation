```python
folder_path = path1
save_path = '../../results/experimental_plan_fm_update/region/integrate'
os.makedirs(save_path, exist_ok=True)

# List all relevant CSV files
all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

descriptors = set()
for f in all_files:
    if f.startswith('low_'):
        descriptor = f.replace('low_', '').replace('_dispense_volumes.csv', '')
        descriptors.add(descriptor)

combined_dfs = {}

for descriptor in descriptors:
    low_file = f"low_{descriptor}_dispense_volumes.csv"
    high_file = f"high_{descriptor}_dispense_volumes.csv"

    df_low = pd.read_csv(os.path.join(folder_path, low_file))
    df_high = pd.read_csv(os.path.join(folder_path, high_file))

    combined_df = pd.concat([df_low, df_high], ignore_index=True)
    combined_dfs[descriptor] = combined_df

    output_file = f"combined_{descriptor}_dispense_volumes.csv"
    combined_df.to_csv(os.path.join(save_path, output_file), index=False)
    print(f"Saved: {output_file}")
```

```python
import os
import pandas as pd

# Set the folder containing the CSV files
folder_path = path2
save_path = '../../results/experimental_plan/special/integrate'

files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Reference file to get preferred column order
ref_file = 'sample_special1_dispense_volumes.csv'
ref_df = pd.read_csv(os.path.join(folder_path, ref_file), sep='\t' if ref_file.endswith('.tsv') else ',')

# Get all unique columns from all files
all_columns = set(ref_df.columns)
for file in files:
    df = pd.read_csv(os.path.join(folder_path, file), sep='\t' if file.endswith('.tsv') else ',')
    all_columns.update(df.columns)

# Preserve reference column order and add any new ones at the end
ordered_columns = list(ref_df.columns) + sorted(all_columns - set(ref_df.columns))

# Collect all aligned rows
combined_rows = []
for file in files:
    df = pd.read_csv(os.path.join(folder_path, file), sep='\t' if file.endswith('.tsv') else ',')
    row = df.iloc[0] if not df.empty else pd.Series(dtype=float)
    row_dict = row.to_dict()
    aligned_row = {col: row_dict.get(col, 0.0) for col in ordered_columns}
    aligned_row['source_file'] = file
    combined_rows.append(aligned_row)

final_df = pd.DataFrame(combined_rows)
output_file = f"combined_special_sample_dispense_volumes.csv"
final_df.to_csv(os.path.join(save_path, output_file), index=False)
```
