```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from chemeleon import chemeleon_latent_space, build_mixture_latent_features
```

```python
df_mixture = pd.read_csv('../../data/df_mixture_20250502.csv')
df_mixture_norm = df_mixture.copy()

x_latent, y = build_mixture_latent_features(
    df=df_mixture_norm,
    smi_cols=['smi1', 'smi2'],
    conc_cols=['conc1', 'conc2'],
    target_col='miscibility',
    latent_fn=chemeleon_latent_space
)

df_total = pd.concat([pd.DataFrame(x_latent),pd.DataFrame(y, columns=['miscibility'])], axis=1)
```

```python
df_total.to_csv('../../data/weight_function_data/df_total_binary_chemeleon_wf_ori.csv', index=False)
```
